```python
#Dictionary
#1.Create an empty dictionary called dog
dog_dictionary = {}
print("Empty dog dictionary....\n dog::",dog_dictionary,"\n")                            #empty dog dictionary

#2.Add name, color, breed, legs, age to the dog dictionary
dog_dictionary = {
        'name':'shibu',
        'color':'white',
        'breed':'shitzu',                                                      #adding keys and values to dog dictionary
        'legs':'4',
        'age':'3'
      }   

print("Printing Dog dictionary.....\n dog::",dog_dictionary,"\n")

#3.Create a student dictionary and add first_name, last_name, gender, age, marital status, 
#  skills, country, city and address as keys for the dictionary

student_dictionary={
    'first_name':'sravanti',
    'last_name':'cherukuri',
    'gender':'female',
    'age':'30',
    'marital status':'married',                                                 #adding keys and values to student dictionary  
    'skills':['Java','SpringBoot','MySQL','MongoDB','AWS'],
    'country':'USA',
    'city':'kansas',
    'address':{
                'street':'131 st Terrace',
                'postalCode':'66213'     
              }
}
print("Printing Student dictionary.....\n student::",student_dictionary,"\n")

# 4.Get the length of the student dictionary
print("The Length of Student Dictionary is ::",len(student_dictionary),"\n")                #finding length of student dictionary


# 5.Get the value of skills and check the data type, it should be a list
print("Value of skills from student dictionary #Way1::",student_dictionary.get('skills'),"\n")    #fetching skills property
print("Value of skills from student dictionary #Way2::",student_dictionary['skills'],"\n")                                                                                  #using get(key){key='skills'} 
    
print("Data type of skills #Way1::",type(student_dictionary.get('skills')),"\n")              #finding data type of skills property
print("Data type of skills #Way2::",type(student_dictionary['skills']),"\n")

# 6.Modify the skills values by adding one or two skills
student_dictionary['skills'].append('Devops')                                                #adding skill to skills list
student_dictionary['skills'].append('Angular')

print("Value of modified skills from student dictionary::",student_dictionary.get('skills'))

#Get the dictionary keys as a list

print("Student dictionary keys::",student_dictionary.keys())                       #printing student Dictionary keys

#Get the dictionary values as a list

print("Student dictionary values::",student_dictionary.values())                   #printing student Dictionary values



    
```
