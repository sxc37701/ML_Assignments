```python
it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}

print("IT companies set...",it_companies)

#Find the length of the set it_companies
print("The length of the set it_companies ::",len(it_companies))       #finding length of it_companies set

#Add 'Twitter' to it_companies
it_companies.add('Twitter')                                          #adding 'Twitter' to set of IT companies

print("IT companies set...",it_companies)

#Insert multiple IT companies at once to the set it_companies
it_companies.update(['Anthem','Wipro','Dell'])                       #adding multiple items into list

print("IT companies set...",it_companies)

#Remove one of the companies from the set it_companies

it_companies.remove('Wipro')                                          #removing one element from set

print("IT companies set...",it_companies)

#What is the difference between remove and discard

#it_companies.remove('HP')                                            #attempt to remove non existing element results in 'key error'
#print("IT companies set...",it_companies)

it_companies.discard('HP')                                            #attempt to remove non existing element 
print("IT companies set...",it_companies)                             #doest not throw any error

#Join A and B
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}

C = A.union(B)                                                         #joining A and B sets
print("Union of Set A and Set B...",C) 

#Find A intersection B
print("Intersection of Set A and Set B",A.intersection(B))             #intersection of A and B

#Is A subset of B
print("Is Set A subset of Set B",A.issubset(B))                        #checking if A subset of B


#Are A and B disjoint sets
print("Are Set A and Set B disjoint sets",A.isdisjoint(B))              #checking if A & B are disjoint sets

#Join A with B and B with A
A.update(B)
print("Joining A with B",A)                                             #joining A with B
B.update(A)
print("Joining B with A",B)                                             #joining B with A
#What is the symmetric difference between A and B
print("Symmetric difference between A and B",A.symmetric_difference(B))  #finding symmetric difference between A and B
#Delete the sets completely
del it_companies                                                         #deleting sets A B and it_companies
del A
del B

#print(it_companies)

#Convert the ages to a set and compare the length of the list and the set.
age = [22, 19, 24, 25, 26, 24, 25, 24]
set_age = set(age)
print("Age set ::",set_age)
print("Comparing lengths of set and list",len(age)>len(set_age))    




```
